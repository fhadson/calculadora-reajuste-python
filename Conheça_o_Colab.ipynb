```
import math
import time

angulo= float(input('Digite o angulo que voce deseja:'))
seno= math.sin(math.radians(angulo))
cossen= math.cos(math.radians(angulo))
tang= math.tan(math.radians(angulo))

tamanho = 20
print('carregando o calculo. . .')
for i in range(tamanho +1):
  preenchido = ('█'* i)
  vazio = '-' * (tamanho - i)
  porcentagem =(i/ tamanho) * 100
  print(f'\rCarregando: |{preenchido}{vazio}| {porcentagem:.0f}%', end=" ")
  time.sleep(0.1)

print(f'\r| O angulo de seno de  {angulo} é: {seno :.2f}')
print(f"| O ângulo de cosseno é de: {cossen :.2f} ")
print(f'| O ângulo de tangente é de: {tang :.2f}')
```
